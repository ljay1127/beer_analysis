```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('beer_data_set.csv')
```

```python
y = ['Ave Rating']
x = ['ABV', 'Min IBU', 'Max IBU', 'Astringency', 'Body',
       'Alcohol', 'Bitter', 'Sweet', 'Sour', 'Salty', 'Fruits', 'Hoppy',
       'Spices', 'Malty']
```

```python
scaler = MinMaxScaler(feature_range=(1,100))
df[x] = scaler.fit_transform(df[x])
```

```python
plt.figure(figsize=(30,10))
for i in x:
    plt.scatter(df[i], df[y], label=i)
plt.title('Beer Plot')
plt.ylabel('Rating')
plt.legend()
plt.show()
```
